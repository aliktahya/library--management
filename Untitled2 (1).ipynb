```python
class Library:
    def __init__(self, filename="books.txt"):
        self.filename = filename
        self.file = open(self.filename, "a+")

    def __del__(self):
        self.file.close()
        print(f"File '{self.filename}' has been closed.")


    def list_books(self):
        self.file.seek(0)
        book_lines = self.file.readlines()
        if book_lines:
            print("List of Books:")
            for line in book_lines:
                book_info = line.strip().split(',')
                title, author = book_info[:2]
                print(f"Title: {title}, Author: {author}")
        else:
            print("No books available in the library.")

    def add_book(self):
        title = input("Enter the title of the book: ")
        author = input("Enter the author of the book: ")
        year = input("Enter the first release year of the book: ")
        pages = input("Enter the number of pages of the book: ")
        book_info = f"{title},{author},{year},{pages}\n"
        self.file.write(book_info)
        print(f"Book '{title}' has been added to the library.")

    def remove_book(self):
        title_to_remove = input("Enter the title of the book to remove: ")
        self.file.seek(0)
        book_lines = self.file.readlines()
        updated_books = []
        found = False
        for line in book_lines:
            if title_to_remove in line:
                found = True
                continue
            updated_books.append(line)
        if not found:
            print(f"Book '{title_to_remove}' not found in the library.")
            return
        self.file.seek(0)
        self.file.truncate()
        for book in updated_books:
            self.file.write(book)
        print(f"Book '{title_to_remove}' has been removed from the library.")

def show_menu():
    while True:
        menu = """*** MENU ***
        1) List Books
        2) Add Book
        3) Remove Book
        4) Exit
        """
        menu_choice = input("Enter your choice (1/2/3/4): ")

        if menu_choice == '1':
            lib.list_books()
        elif menu_choice == '2':
            lib.add_book()
        elif menu_choice == '3':
            lib.remove_book()
        elif menu_choice == '4':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please select 1, 2, 3, or 4.")

lib = Library()
show_menu()
```
